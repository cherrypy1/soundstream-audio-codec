# 4. Download sound sample

```python
from pathlib import Path
from urllib.request import urlretrieve

URL = "https://keithito.com/LJ-Speech-Dataset/LJ025-0076.wav"  # URL для выбранной записи
 
Path("outputs").mkdir(exist_ok=True)
input_path = "outputs/demo_input.wav"
output_path = "outputs/demo_reconstructed.wav"

urlretrieve(URL, input_path)  # Загрузка семпла по пути input_path
input_path
```

# 5. Model generation

```python
from inference import resynthesize_sound

resynthesize_sound("checkpoints/final.ckpt", input_path, output_path) # Функция для инференса модели
output_path
```

# 6. Result audio

```python
from IPython.display import Audio, display

print("Original")
display(Audio(input_path)) # Показ исходного аудио

print("Reconstructed")
display(Audio(output_path))  # Показ реконструированного аудио
```
